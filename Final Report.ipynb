# 1.2 Methodology 

Our overall approach and methodology can be summarized using this image.



To complete our analysis, first we had to process our data. One of the challenge that we faced was that we had multiple different sources of our data, resulting in the format of the data being different to combine for analyses. Another challenge we faced was missing values in our data. Our data was three dimensional with more than 180 countries with multiple metrics measured around tens of years.

Due to the above factors, we as a team decided to shorlist the data for the purpose of our analysis. We are considering years from 2008 till 2017 for countries Bulgaria, Brazil, Chile, China, Spain, India, Jordan, Kenya, Mexico, Mali, Nepal, New Zealand, Qatar, Rwanda, Senegal, Singapore, El Salvador and Ukraine. 

The selection of countries has been made to make the data fully representative. The countries represent the distribution of each province.

Our Structured tables along with their relationships with each other are as follows:


#### Question 1: **Does the increase in the quality of education improve social and economical welfare of people (HDI Index)?**

First of all, we wanted to establish the relationship between the Education Quality Index(EQI) and the Human Development Index (HDI). We examined this relationship by using data for the year 2017 (the most recent data) to plot each country with EQI as X-axis and HDI on Y-Axis. Then we ran a linear regression between both variables to examine the relationship and the trend.

Secondly, we looked at the same relationship as above but grouped and faceted into by country's income level. We did this to see if the correlation differs between different countries by income level.

Thirdly, we looked at the same relationship (EDI vs HDI) but grouped and faceted by geographical regions of countries to see if the correlation differs between different regions. The regions are the same as Q1 i.e. East Asia & Pacific, Middle East & North Africa, Latin America & Caribbean, Europe and Central Asia, SUb-Saharan Africa and South Asia.

Finally, we attemplt to examine the long term relationship between Quality of Education and Human Development Index by Income group Between 2008 and 2017. This is again a **Time Series** and longitudinal analysis which shows us the trend of how with each passing year, HDI changes by change in quality of education. 

#### Question 2: **Does the increase in quality of education result in an increase in the living standards of the people (GDP)?**

To understand the relationship, we plotted GDP per Capita vs. Quality of Education Index in 2017. While plotting, we used the **Log** of GDP per capital to rescale the data and make it more interpretable. We have Quality of Education Index on X-Axis and GDP per Capita on Y-axis. 
On the scatter plot, we drew a OLS Trend Line to see how strong the relationship between both the variables is. We want to see that as our Quality of education increases, does living stands of people (GDP per capita) also increase or not?

In the next visualization, we carry out a **longitudinal and time series analysis** for each country, to see if there is a long term impact on the GDP per capita with change in quality of education.

#### Question 3: **Does the education expenditure as a % of the GDP correlate with increase in quality of education?**

Now, since we established the relationship between quality of education and socioeconomic welfare, we want to uncover the factors that would help policy makers improve the quality of education. Accordingly, we looked at the overall relationship between **education expenditure as a % of the GDP** against **Education Quality Index** for 2017 (the most recent data). We have plotted each country with EQI as Y-axis and Education spend as % of GDP on X-Axis. Then we ran a Simple linear regression to identify the corelation between the 2 variables. 

Secondly, we tried to slice the data by both the country's income level and the region as defined by the worldbank to uncover any underlying trends. We saw some interesting insights which will be shared in the **results **

Finally, we tried to look at the long-term effects through the visualization "Relationship between Budget, GDP and Education Quality by Country" which  gives a **Time Series** analysis of how with passing years, _quality of education_ changes with change in education as a percentage of GDP and Budget.

#### Question 4: **Do scores in international standardized tests provide a good indicator of the quality of education?**

While the Education Quality Index is adopted by the World Economic Forum, the World Bank and a multitude of different organizations, it carries a lot of subjectivity as it relies on questionnaires. So we attempt to validate the results through examining the relationship with quantitative measure of **PISA** test scores.

To examine this relationship, we tried to observe which countries take part in the international test score **PISA**. The visualization plots the countries colored by their income-level on the globe. Here, we clearly see only high and upper middle income countries take part in this test. Which kind of shows the disparity between education standards by income.

The data for PISA scores was availabe subject wise that is PISA Science, Maths and Reading scores. We first plotted the correlation between Science score, Maths score and Reading score over years. As the relationship was highly correlated (r= 0.93), we combined the Pisa subject scores to come up with a single score without losing any trends, which be will used in further analysis.

In the end, we plotted PISA combined scores against the Quality of education. And looked at how the relationship between them have been changing with time. We linear OLS trend line tells us the relationship and how strong it is.

# 1.3 Results

### Question 1: Does the increase in the quality of education improve social and economical welfare of people (HDI Index)?

**At A Glance: Yes**. There seems to be a positive associated between Quality of Education and HDI. The R^2 of 53% suggests that at least 53% of the variation in HDI is explained by the quality of education. A positive slope of our trendline suggests that as the Quality of Education increases, the Social and Economic condition (HDI) also increases.

**Deep Dive**: We see similar trends when we deep dive into our analysis by Country Income levels. We see strong relationship in High Income and Low Income countries. The trend is weak in middle and upper midle income countries. 

If we cross-sectionally analyse the variables by Geographical Region, we again see strong positive relationship. This relationship is particularly strong in East Asia and Pacific as well as Europe and Central Asia.

Our deep dive confirms our conclusion mentioned in the beginning.

### Question 2: Does the education expenditure as a % of the GDP correlate with increase in quality of education?

**At A Glance: No.** Based on the analysis performed, it is safe to say that _Education Expenditure as a % of the GDP_ **DOES NOT** have a strong relationship with _Quality of Education_. In other words, the change in Quality of Education cannot be explained by change in Educational spend. We see R^2 value of 6% which suggests this conclusion of a weak relationship. 

**Deep Dive**: If we deep dive into the same relationship by Country Income Group, we still see low R^2 values for each income group. As a generic trend, we do see that on average countries with high income have a greater quality of education compared to the rest. We cannot say much about the trend line as its a weak relationship with high squared error.


If we see our target variables with respect to Geographical Region, apart from Africa, we start seeing some relationship.

We analyze the strength of the relationship, we refer to the R^2 values of the OLS trendline.


### Question 3: Does the increase in quality of education result in an increase in the living standards of the people (GDP per Capita)?

**At A Glance: Yes**. 55% of the increase in GDP per Capita (or living standards of citizen) can be explained by their Quality fo Education.

### Question 4: Do scores in international standardized tests provide a good indicator of the quality of education?

We can say that the test scores are a fairly decent indicator of quality of education. Althouh we have a high R^2 of 66%, we still have 34% of the variation which is explored by some other variables (not in scope of this study).

# 1.4 Conclusion


While we started our analysis seeking to examine the relationship between the quality of education and the socio-economic welfare, we stumbled upon much more important observations. Yes there's a positive relationship between the quality of education and the socio-economic welfare as analyzed using the EQI and HDI. ### Do scores in international standardized tests provide a good indicator of the quality of education?

We will use PISA as an example of the international standardized tests. Our objective is to analyze if students score well on those tests because they are trained to perform well in such standardized test settings or because they are actually getting a high education quality. 

```python
# Loading the data from db
# Main returns: PISA Math Scores (pisamath), PISA Science Scores (pisasci), PISA Reading Scores (pisaread), Education Quality Index(val_qual_ed)
# Supplementary information: Country Code, Country Name, Year, Income Group , Region
qu_scores_qual = '''SELECT ct.country_code, ct.country_name, ps.year, ps.pisamath,ps.pisasci, ps.pisaread, val_qual_ed, ct.income_grp, ct.region
FROM edx.education_indicators as ei 
	INNER JOIN edx.pisa_scores as ps ON ei.year = ps.year and ei.country_code = ps.country_code 
	INNER JOIN edx.country as ct on ct.country_code  = ei.country_code
WHERE val_qual_ed is not NULL and ps.year BETWEEN 2000 and 2020 '''

cur.execute(query=qu_scores_qual)
df_scores_qual = pd.DataFrame(cur.fetchall())

df_scores_qual = df_scores_qual.dropna()

df_scores_qual.columns = ['country_code','country_name', 'year', 'pisa_math','pisa_science','pisa_reading','val_qual_ed','income_grp','region']

df_scores_qual
```

Which countries are particpating in the PISA test?

```python
fig = px.choropleth(locations=df_scores_qual['country_code'], 
                    locationmode="ISO-3",
                    color=df_scores_qual['income_grp'],
                    color_discrete_map= color_dict_4,
                    scope="world",
                    labels={'color':'Income Group'}
                   )

fig.show()
```

We can see from the graph that participants of PISA are generally high income and upper middle income ... !!!! INEQUALITY!!!!

Let's see if there are any trends in the PISA scores.

```python
fig = px.scatter(data_frame=df_scores_qual,x='pisa_math',y='pisa_science',facet_col='year',hover_name='country_name',color='pisa_reading',trendline='ols', facet_col_wrap=3, labels={'pisa_reading':'PISA Reading'},
                 color_continuous_scale=px.colors.sequential.Teal, trendline_color_override='Purple')
                 

#Setting the labels of the plot
fig.update_layout(
    title={
        'text': 'Correlation among PISA categories scores',
        'x':0.5,
        'xanchor': 'center'},
    xaxis_title='PISA Math Score',
    yaxis_title='PISA Science Score'
)

#Ensuring that all x-axis show the correct label
fig.for_each_xaxis(lambda x: x.update(title="PISA Math Score"))

fig.show()
```

Ok the scores are highly correlated R2 of 0.93 with each other throughout all the testing years. So it is safe to combine them without losing any trends.

We'll create 1 metric that combines all the scores together to perform our analysis.

```python
df_scores_qual['pisa_combined_score'] = df_scores_qual.apply(lambda row: row['pisa_math'] + row['pisa_science'] + row['pisa_reading'],axis=1)
df_scores_qual
```

```python
fig = px.scatter(data_frame=df_scores_qual,x='pisa_combined_score',y='val_qual_ed',facet_col='year',hover_name='country_name',color='val_qual_ed',trendline='ols', labels={'val_qual_ed':'Education Quality Index'},
                 color_continuous_scale=px.colors.sequential.Teal, trendline_color_override='Purple')

#Setting the labels of the plot
fig.update_layout(
    title={
        'text': 'Relationship between PISA combined score vs. Quality of Education Index',
        'x':0.5,
        'xanchor': 'center'},
    xaxis_title='PISA Combined Score',
    yaxis_title='Quality of Education Index',
    legend={'title':'Quality of Education Index'}
)
#Ensuring that all x-axis show the correct label
fig.for_each_xaxis(lambda x: x.update(title="PISA Combined Score"))

fig.show()
```
