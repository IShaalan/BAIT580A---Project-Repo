```python
#Transposing indicators as columns and years as rows for GDP table
cur.execute("""CREATE TABLE  edx.gdp as(select country_name,country_code, year
,sum(gdp_current_us)as gdp_current_us
,sum(gdp_growthperc) as gdp_growthperc
from
(select country_name,country_code,year
,case when series_name in ('GDP (current US$)') then value end as GDP_current_US
,case when series_name in ('GDP growth (annual %)') then value end as GDP_growthperc
from

(SELECT c.country_name,c.country_code,c.series_name,c.series_code,t.* 
FROM edx.test1_gdp c

cross join lateral (
values 
(c.y_2001, '2001'),
(c.y_2002, '2002'),
(c.y_2003, '2003'),
(c.y_2004, '2004'),
(c.y_2005, '2005'),
(c.y_2006, '2006'),
(c.y_2007, '2007'),
(c.y_2008, '2008'),
(c.y_2009, '2009'),
(c.y_2010, '2010'),
(c.y_2011, '2011'),
(c.y_2012, '2012'),
(c.y_2013, '2013'),
(c.y_2014, '2014'),
(c.y_2015, '2015'),
(c.y_2016, '2016'),
(c.y_2017, '2017'),
(c.y_2018, '2018'),
(c.y_2019, '2019'),
(c.y_2020, '2020')
)as t(value,year)
 )as ac)
 as ab
 group by 1,2,3
 order by 1,3
 )""")
 
conn.commit()
 
```
